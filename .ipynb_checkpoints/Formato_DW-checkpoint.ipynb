```python
df['precio'] = df['precio'].str.replace('$', '').str.replace('.', '').astype(float)
df['promo'] = df['promo'].str.replace('$', '').str.replace('.', '').astype(float)
df['marca'] = df['marca'].astype('string')
df['descripcion'] = df['descripcion'].astype('string')
df['imagen'] = df['imagen'].astype('string')
df['promo_exclusiva'] = df['promo_exclusiva'].astype('string')
df['link'] = df['link'].astype('string')
```

```python
df['promocion'] = np.where(df['precio'].isna(), 'SIN PROMO', 'CON PROMO')
df['precio'] = df['precio'].fillna(df['promo'])
```

```python
def extract_ml(description):
    # Buscamos una cantidad seguida de "ml" o "L"
    match = re.search(r'(\d+(\.\d+)?)[\s-]?([mMlL])', description)
    if match:
        quantity, _, unit = match.groups()
        quantity = float(quantity)
        # Convertimos a ml si es necesario
        if unit.lower() == 'l':
            quantity *= 1000
        return int(quantity)
    return None  # Devolvemos None si no encontramos una cantidad

# Aplicamos la función a cada descripción
df['Cantidad (ml)'] = df['descripcion'].apply(extract_ml)
df['Cantidad (ml)'] = df['Cantidad (ml)'].astype('Int64', errors='ignore')

# Ahora df tiene una nueva columna 'Cantidad (ml)' con la cantidad de ml extraída de la descripción
```

```python
def extract_fragrance(description):
    # Dividir la descripción en palabras
    words = description.split()
    
    # Buscar la posición donde termina el nombre de la marca y comienza la fragancia
    # En tus ejemplos, parece que la fragancia comienza después de la primera o segunda palabra
    start_pos = 2
    
    # Buscar la posición donde comienza la información de tamaño/volumen
    # En tus ejemplos, parece que esta información comienza en la última o penúltima palabra
    end_pos = len(words) - 2
    
    # Unir las palabras entre start_pos y end_pos para formar la fragancia
    fragrance = ' '.join(words[start_pos:end_pos])
    
    return fragrance

# Suponiendo que df es tu DataFrame y la columna con la descripción se llama 'descripcion'
df['Fragancia'] = df['descripcion'].apply(extract_fragrance)
```

```python
import re
import pandas as pd

def process_promo_exclusiva(row):
    value = row['promo_exclusiva']
    precio_original = row['precio']
    
    if pd.isna(value):
        return '0%'
    
    # Para el caso "30% dcto."
    pct_match = re.match(r'(\d+)%', value)
    if pct_match:
        return f'{pct_match.group(1)}%'
    
    # Para el caso "3 por 2 a $8.518"
    three_for_two_match = re.match(r'3 por 2 a \$(\d+)', value)
    if three_for_two_match:
        return '33.3%'
    
    # Para el caso "Lleva 2 por $2.179"
    two_for_price_match = re.match(r'Lleva 2 por \$([\d\.]+)', value)
    if two_for_price_match:
        # Eliminar los puntos de los miles antes de convertir a float
        precio_promocional_str = two_for_price_match.group(1).replace('.', '')
        precio_promocional_por_unidad = float(precio_promocional_str) / 2
        descuento_pct = (1 - (precio_promocional_por_unidad / precio_original)) * 100
        return f'{descuento_pct:.1f}%'
    
    # Para el caso "Paga $1.299"
    pay_match = re.match(r'Paga \$(\d+\.\d+|\d+)', value)
    if pay_match:
        # Eliminar los puntos de los miles antes de convertir a float
        precio_promocional_str = pay_match.group(1).replace('.', '')
        precio_promocional = float(precio_promocional_str)
        descuento_pct = (1 - (precio_promocional / precio_original)) * 100
        return f'{descuento_pct:.1f}%'
    
    return 'Desconocido'

# Aplicar la función a las filas del DataFrame, pasando toda la fila a la función
df['PORCENTAJE_DESCUENTO_EXCLUSIVA'] = df.apply(process_promo_exclusiva, axis=1)
```

```python
##Imágenes

# Definir la ruta de la carpeta donde deseas guardar las imágenes
carpeta_imagenes = 'C:/Users/man27/Desktop/AI_test/UL_DW/Imagenes'  # Cambia esto a la ruta que desees

def descargar_guardar_imagen(row):
    url = row['imagen']  # Cambiado a 'imagen' para coincidir con tu DataFrame
    marca = row['marca']
    formato = str(row['Cantidad (ml)'])  # Usando 'Cantidad (ml)' para el formato, convertido a string
    fragancia = row['Fragancia']
    # Reemplazando posibles caracteres no deseados en los nombres de los archivos
    nombre_archivo = f"{marca}_{formato}_{fragancia}.jpg".replace(" ", "_").replace("/", "_").replace("<NA>", "NA").replace("..", ".")
    # Crear la ruta completa donde se guardará la imagen
    ruta_archivo = os.path.join(carpeta_imagenes, nombre_archivo)
    # Descargar y guardar la imagen
    respuesta = requests.get(url)
    if respuesta.status_code == 200:  # Verificar que la solicitud fue exitosa
        with open(ruta_archivo, 'wb') as archivo:
            archivo.write(respuesta.content)
    else:
        print(f'No se pudo descargar la imagen de {url}')

# Asegúrate de que la carpeta existe, y si no, créala
os.makedirs(carpeta_imagenes, exist_ok=True)

# Aplicar la función a cada fila del DataFrame
df.apply(descargar_guardar_imagen, axis=1)
```
